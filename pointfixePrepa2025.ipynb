```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ============================================
# 1. DÉFINIR LA FONCTION (À MODIFIER ICI)
# ============================================
def f(x, y):
    """
    Fonction de point fixe: (x, y) -> (f1(x,y), f2(x,y))
    Exemple: système x = cos(y), y = sin(x)
    """
    x_new = np.cos(y)
    y_new = np.sin(x)
    return x_new, y_new
```

```python

# ============================================
# 2. PARAMÈTRES
# ============================================
x0, y0 = 0.0, 0.0  # Point initial
n_iterations = 15   # Nombre d'itérations
```

```python
# ============================================
# 3. ITÉRATIONS ET TABLEAU
# ============================================
x_vals = [x0]
y_vals = [y0]

print("="*60)
print(f"{'k':<5} {'x_k':<12} {'y_k':<12} {'erreur':<12}")
print("="*60)

for k in range(n_iterations):
    x_k, y_k = x_vals[-1], y_vals[-1]
    x_new, y_new = f(x_k, y_k)

    # Calcul de l'erreur
    erreur = np.sqrt((x_new - x_k)**2 + (y_new - y_k)**2)

    x_vals.append(x_new)
    y_vals.append(y_new)

    print(f"{k:<5} {x_k:<12.6f} {y_k:<12.6f} {erreur:<12.6f}")

# Afficher le dernier point (approximation du point fixe)
print("="*60)
print(f"Point fixe approché: x* ≈ {x_vals[-1]:.6f}, y* ≈ {y_vals[-1]:.6f}")
print("="*60)
```

```python
# ============================================
# 4. REPRÉSENTATION GRAPHIQUE
# ============================================
plt.figure(figsize=(10, 8))

# Tracer les points
plt.plot(x_vals, y_vals, 'bo-', linewidth=2, markersize=8, label='Itérations')

# Point initial en vert
plt.plot(x0, y0, 'go', markersize=12, label='Point initial', zorder=5)

# Point fixe (dernier point) en rouge
plt.plot(x_vals[-1], y_vals[-1], 'r*', markersize=20, label='Point fixe approché', zorder=5)

# Annoter les premiers points
for i in range(min(5, len(x_vals))):
    plt.annotate(f'$x_{i}$', (x_vals[i], y_vals[i]),
                 textcoords="offset points", xytext=(5,5), fontsize=10)

# Grille et labels
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Itération de point fixe dans $\mathbb{R}^2$', fontsize=16)
plt.legend(fontsize=12)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Fonction et paramètres
def f(x, y):
    return np.cos(y), np.sin(x)

# Itérations de point fixe
x0, y0 = 0.0, 0.0
n_iter = 10
x_vals, y_vals = [x0], [y0]

for k in range(n_iter):
    x_new, y_new = f(x_vals[-1], y_vals[-1])
    x_vals.append(x_new)
    y_vals.append(y_new)

# ============================================
# FIGURE COMBINÉE
# ============================================
fig = plt.figure(figsize=(18, 6))

# 1. Surface f1(x,y) = cos(y)
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z1 = np.cos(Y)

ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='viridis', alpha=0.7, edgecolor='none')
ax1.plot(x_vals, y_vals, [np.cos(y_vals[i]) for i in range(len(y_vals))],
         'ro-', linewidth=3, markersize=8)
ax1.set_title('$f_1(x,y) = \cos(y)$', fontsize=14)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')

# 2. Surface f2(x,y) = sin(x)
Z2 = np.sin(X)

ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X, Y, Z2, cmap='plasma', alpha=0.7, edgecolor='none')
ax2.plot(x_vals, y_vals, [np.sin(x_vals[i]) for i in range(len(x_vals))],
         'ro-', linewidth=3, markersize=8)
ax2.set_title('$f_2(x,y) = \sin(x)$', fontsize=14)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')

# 3. Trajectoire dans le plan (x,y)
ax3 = fig.add_subplot(133)
ax3.plot(x_vals, y_vals, 'bo-', linewidth=2, markersize=8)
ax3.plot(x0, y0, 'go', markersize=12, label='Départ')
ax3.plot(x_vals[-1], y_vals[-1], 'r*', markersize=20, label='Point fixe')
ax3.grid(True, alpha=0.3)
ax3.set_xlabel('x', fontsize=12)
ax3.set_ylabel('y', fontsize=12)
ax3.set_title('Itérations dans le plan', fontsize=14)
ax3.legend()
ax3.axis('equal')

plt.tight_layout()
plt.show()
```
